```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/world_population.csv", index_col='rank')
df.head()
```

```python
query = """
CREATE TABLE IF NOT EXISTS worldpop(
  rank INT,
  country TEXT PRIMARY KEY,
  density DOUBLE,
  population DOUBLE NOT NULL,
  area INT NOT NULL
);
"""
cursor.execute(query)
db.commit()
```

```python
uz = df[df.country == 'Uzbekistan']
uz
```

```python
rank =uz.index.values[0]
country = uz.at[rank, 'country']
density = uz.at[rank, 'density']
population = uz.at[rank, 'pop2021']
area = uz.at[rank, 'area']
print(f"{rank}, {country}, {density}, {population}, {area}")
```

```python
query = f"""
INSERT INTO worldpop VALUES ({rank}, '{country}', {density}, {population}, {area})
"""
cursor.execute(query)
db.commit()
```

```python
import pandas as pd
import numpy as np
import sqlite3 as sql
from pprint import pprint as print
```

```python
!wget "https://github.com/anvarnarz/praktikum_datasets/raw/main/electricity.db"
```

```python
db = sql.connect("electricity.db")
cursor = db.cursor()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

```python
cursor.execute("PRAGMA table_info(elect);")
cursor.fetchall()
```

```python
pd.read_sql("SELECT * FROM elect LIMIT 15", db, index_col='index')
```

```python
query = "SELECT DISTINCT country FROM elect"
countries = pd.read_sql(query, db)
```

```python
'Uzbekistan' in countries['country'].values
```

```python
query = "SELECT * FROM elect WHERE country='Uzbekistan' AND year>2010"
dfuz = pd.read_sql(query, db)
dfuz.head()
```

```python
query = "SELECT country, year, coal, gas, hydro FROM elect WHERE country='Uzbekistan' AND year>2010"
dfuz = pd.read_sql(query, db)
dfuz.head()
```

```python
#BETWEEN
query = "SELECT country, year, coal, gas, hydro FROM elect WHERE country='Uzbekistan' AND year BETWEEN 2005 AND 2015"
pd.read_sql(query, db)
```

```python
#IN
query = "SELECT country, year, coal, gas, hydro FROM elect WHERE country='Uzbekistan' AND year IN(1990, 1995, 2000, 2005)"
pd.read_sql(query, db)
```

```python
#LIKE
# %var - Anvar, Sarvar
# _- _ar -> car, bar, tar,
# __ar - star, scar
# %ol% - bola, lola, bolalar

query = "SELECT country, year, coal, gas, hydro FROM elect WHERE country LIKE'%stan' AND year IN(2019, 2020, 2021)"
pd.read_sql(query, db)
```

```python
query = "SELECT * FROM elect WHERE code LIKE'A__' AND year=2020"
pd.read_sql(query, db)
```
